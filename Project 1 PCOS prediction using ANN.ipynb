```python
X = df.iloc[:, 1:].values
y = df.iloc[:, 0].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#ANN model
import tensorflow as tf
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=100, activation='relu'))
ann.add(tf.keras.layers.Dense(units=500, activation='relu'))
# ann.add(tf.keras.layers.Dropout(0.20))
ann.add(tf.keras.layers.Dense(units=1000, activation='relu'))
# ann.add(tf.keras.layers.Dense(units=10, activation='relu'))
# ann.add(tf.keras.layers.Dropout(0.20))
ann.add(tf.keras.layers.Dense(units=500, activation='relu'))
# ann.add(tf.keras.layers.Dropout(0.20))
# ann.add(tf.keras.layers.Dense(units=10, activation='relu'))
ann.add(tf.keras.layers.Dense(units=10, activation='relu'))
# ann.add(tf.keras.layers.Dropout(0.40))
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
#activation function for non-binary classification is softmax
```

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
#loss for non-binary classification is 'categorical_crossentropy'
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping
es = EarlyStopping(monitor='accuracy', min_delta=0.01, patience=3, verbose=1, mode='auto')
mc = ModelCheckpoint(monitor='accuracy', filepath = "D:\Projects\Datasets\PCOS_data\Best_model.h5", verbose=1, save_best_only = True, mode='auto')
cd = [es, mc]
```

```python
ann.fit(X_train, y_train, batch_size = 32, epochs = 100, steps_per_epoch=10, verbose = 1)
```

```python
y_pred = ann.predict(X_test)
```

```python
y_pred
```

```python
y_pred = (y_pred > 0.5)
```

```python
print(np.concatenate((y_pred.reshape(len(X_test), 1), y_test.reshape(len(X_test), 1)), 1))
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
